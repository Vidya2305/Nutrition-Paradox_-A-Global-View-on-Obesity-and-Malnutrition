```python
special_cases = {
                    'GLOBAL': 'Global',
                    'WB_LMI': 'Low & Middle Income',
                    'WB_HI': 'High Income',
                    'WB_LI': 'Low Income',
                    'EMR': 'Eastern Mediterranean Region',
                    'EUR': 'Europe',
                    'AFR': 'Africa',
                    'SEAR': 'South-East Asia Region',
                    'WPR': 'Western Pacific Region',
                    'AMR': 'Americas Region',
                    'WB_UMI': 'Upper Middle Income'}
```

```python
#3.	Define a function to convert the 3-letter codes to full names
def country_code_converter(code):
    if code in special_cases:
        return special_cases[code]
    else:
        country = pycountry.countries.get(alpha_3=code)
        return country.name
```

```python
#4.	Apply the function to the Country colum
df_obesity['Country'] = df_obesity['Country'].apply(country_code_converter)
df_obesity.head()
```

```python
df_malnutrition['Country'] = df_malnutrition['Country'].apply(country_code_converter)
df_malnutrition.head()
```

```python
#2.	CI_Width: Calculate the confidence interval width.
#■	Formula: CI_Width = High - Low
df_obesity['CI_Width'] = df_obesity['UpperBound'] - df_obesity['LowerBound']
df_obesity.head()
```

```python
df_malnutrition['CI_Width'] = df_malnutrition['UpperBound'] - df_malnutrition['LowerBound']
df_malnutrition.head()
```

```python
"""3.	obesity_level (for the obesity table only):
■	Categorize obesity levels based on NumericValue:
■	>= 30 → High
■	25–29.9 → Moderate
■	< 25 → Low"""
df_obesity['Obesity_level'] = ['High' if value >= 30 else 'Moderate' if 25 <= value < 30 else 'Low' for value in df_obesity['Mean_Estimate']]
df_obesity.head()
```

```python
"""4.	malnutrition_level (for the malnutrition table only):
■	Categorize malnutrition levels based on NumericValue:
■	>= 20 → High
■	10–19.9 → Moderate
■	< 10 → Low"""
df_malnutrition['Malnutrition_level'] = ['High' if value >= 20 else 'Moderate' if 10 <= value < 20 else 'Low' for value in df_malnutrition['Mean_Estimate']]
df_malnutrition.head()
```

```python
#Obesity Table (10 Queries)
#1.	Top 5 regions with the highest average obesity levels in the most recent year(2022)
cursor.execute(
'''
SELECT Region, AVG(Mean_Estimate) AS Avg_Obesity
FROM obesity
WHERE Year = 2022 AND Region IS NOT NULL
GROUP BY Region
ORDER BY Avg_Obesity DESC
LIMIT 5;
'''
)

result1 = cursor.fetchall()
print(pd.DataFrame(result1, columns = ['Region', 'Avg_Obesity']))
```

```python
#2. Top 5 countries with highest obesity estimates
cursor.execute(
'''
SELECT Country, MAX(Mean_Estimate) AS Max_Obesity
FROM obesity
GROUP BY Country
ORDER BY Max_Obesity DESC
LIMIT 5;
'''
)
result2 = cursor.fetchall()
print(pd.DataFrame(result2, columns = ['Country', 'Max_Obesity']))
```

```python
#3.	Obesity trend in India over the years(Mean_estimate)
cursor.execute(
'''
SELECT Year, AVG(Mean_Estimate) AS India_Obesity
FROM obesity
WHERE Country = 'India'
GROUP BY Year
ORDER BY Year;
'''
)
result3 = cursor.fetchall()
print(pd.DataFrame(result3, columns = ['Year', 'Avg_Mean_Estimate']))
```

```python
#4.	Average obesity by gender
cursor.execute(
'''
SELECT Gender, AVG(Mean_Estimate) AS Avg_Obesity
FROM obesity
GROUP BY Gender;
'''
)
result4 = cursor.fetchall()
print(pd.DataFrame(result4, columns = ['Gender', 'Avg_Obesity']))
```

```python
#5.	Country count by obesity level category and age group
cursor.execute(
'''
SELECT Obesity_level, Age_group, COUNT(DISTINCT Country) AS Country_Count
FROM obesity
GROUP BY Obesity_level, Age_group;
'''
)
result5 = cursor.fetchall()
print(pd.DataFrame(result5, columns = ['Obesity_level', 'Age_group', 'Country_Count']))
```

```python
#6.	Top 5 countries least reliable countries(with highest CI_Width) and Top 5 most consistent countries (smallest average CI_Width)
cursor.execute(
'''
SELECT Country, AVG(CI_Width) AS Avg_CI_Width
FROM obesity
GROUP BY Country
ORDER BY Avg_CI_Width DESC
LIMIT 5;
'''
)
result6 = cursor.fetchall()
print(pd.DataFrame(result6, columns = ['Country', 'CI_Width']))
cursor.execute(
'''
SELECT Country, AVG(CI_Width) AS Avg_CI_Width
FROM obesity
GROUP BY Country
ORDER BY Avg_CI_Width ASC
LIMIT 5
'''
)
result61 = cursor.fetchall()
print(pd.DataFrame(result61, columns = ['Country', 'CI_Width']))
```

```python
#7.	Average obesity by age group
cursor.execute(
'''
SELECT Age_group, AVG(Mean_Estimate) AS Avg_Obesity
FROM obesity
GROUP BY Age_group;
'''
)
result7 = cursor.fetchall()
print(pd.DataFrame(result7, columns = ['Age_group', 'Avg_Obesity']))
```

```python
#8.	Top 10 Countries with consistent low obesity (low average + low CI)over the years
cursor.execute(
'''
SELECT Country, AVG(Mean_Estimate) AS Avg_Obesity, AVG(CI_Width) AS Avg_CI
FROM obesity
GROUP BY Country
HAVING Avg_Obesity < 4.75 AND Avg_CI < 2.9
ORDER BY Avg_Obesity ASC
LIMIT 10;
'''
)
result8 = cursor.fetchall()
print(pd.DataFrame(result8, columns = ['Country', 'Avg_Obesity', 'Avg_CI_Width']))
```

```python
#9.	Countries where female obesity exceeds male by large margin (same year)
cursor.execute(
'''
SELECT f.Country, f.Year, f.Mean_Estimate - m.Mean_Estimate AS Difference
FROM obesity f
JOIN obesity m ON f.Country = m.Country AND f.Year = m.Year
WHERE f.Gender = 'Female' AND m.Gender = 'Male' AND (f.Mean_Estimate - m.Mean_Estimate) > 5
ORDER BY Difference DESC;
'''
)
result9 = cursor.fetchall()
print(pd.DataFrame(result9, columns = ['Country', 'Year', 'Difference_between_female_and_male_obesity']))
```

```python
# Combined (5 Queries)
#1. Obesity vs malnutrition comparison by country(any 5 countries)
cursor.execute(
'''
SELECT o.Country, AVG(o.Mean_Estimate) AS Obesity, AVG(m.Mean_Estimate) AS Malnutrition
FROM obesity o
JOIN malnutrition m ON o.Country = m.Country AND o.Year = m.Year
WHERE o.Country IN ('India', 'Nigeria', 'Brazil', 'United States', 'Indonesia')
GROUP BY o.Country;
'''
)
result21 = cursor.fetchall()
print(pd.DataFrame(result21, columns = ['Country', 'Obesity', 'Malnutrition']))
```

```python
#2. Gender-based disparity in both obesity and malnutrition
cursor.execute(
'''
SELECT o.Gender, AVG(o.Mean_Estimate) AS Avg_Obesity, AVG(m.Mean_Estimate) AS Avg_Malnutrition
FROM obesity o
JOIN malnutrition m ON o.Gender = m.Gender AND o.Country = m.Country AND o.Year = m.Year
GROUP BY o.Gender;
'''
)
result22 = cursor.fetchall()
print(pd.DataFrame(result22, columns = ['Gender', 'Avg_Obesity', 'Avg_Malnutrition']))
```

```python
#3.	Region-wise avg estimates side-by-side(Africa and America)
cursor.execute(
'''
SELECT Region, AVG(Mean_Estimate) AS Avg_Estimate, 'Obesity' AS Source
FROM obesity
WHERE Region IN ('Africa', 'Americas')
GROUP BY Region
UNION ALL
SELECT Region, AVG(Mean_Estimate), 'Malnutrition'
FROM malnutrition
WHERE Region IN ('Africa', 'Americas')
GROUP BY Region;
'''
)
result23 = cursor.fetchall()
print(pd.DataFrame(result23, columns = ['Region', 'Avg_Estimate', 'Source']))
```

```python
#4. Countries with obesity up & malnutrition down
cursor.execute(
'''
SELECT o.Country, AVG(o.Mean_Estimate) AS Obesity, AVG(m.Mean_Estimate) AS Malnutrition
FROM obesity o
JOIN malnutrition m ON o.Country = m.Country AND o.Year = m.Year
GROUP BY o.Country
HAVING Obesity > 20 AND Malnutrition < 3
ORDER BY Obesity DESC;
'''
)
result24 = cursor.fetchall()
print(pd.DataFrame(result24, columns = ['Country', 'Obesity', 'Malnutrition']))
```

```python
#5. Age-wise trend analysis
cursor.execute(
'''
SELECT o.Age_group, o.Year, AVG(o.Mean_Estimate) AS Avg_Obesity, AVG(m.Mean_Estimate) AS Avg_Malnutrition
FROM obesity o
JOIN malnutrition m ON o.Age_group = m.Age_group AND o.Country = m.Country AND o.Year = m.Year
GROUP BY o.Age_group, o.Year
ORDER BY o.Age_group, o.Year;
'''
)
result25 = cursor.fetchall()
print(pd.DataFrame(result25, columns = ['Age_group', 'Year', 'Avg_Obesity', 'Avg_Malnutrition']))
```
